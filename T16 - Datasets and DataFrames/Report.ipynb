```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity.
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()` for a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)).

```python
# Compare the average income based on ethnicity
print('The average income of African American is', df[df.Ethnicity == 'African American']['Income'].mean())
print('The average income of Asian is', df[df.Ethnicity == 'Asian']['Income'].mean())
print('The average income of Caucasian is', df[df.Ethnicity == 'Caucasian']['Income'].mean())
```

```python
# On average, do married or single people have a higher balance?
df[['Married', 'Balance']].groupby(['Married']).mean()
```

```python
print('On average single people have a higher balance than married people')
```

```python
# What is the highest income in our dataset?
print('The highest income in the dataset is', df.loc[:, 'Income'].max())
```

```python
# What is the lowest income in our dataset?
print('The lowest income in the dataset is', df.loc[:, 'Income'].min())
```

```python
# How many cards do we have recorded in our dataset?
print('There are', df['Cards'].sum(), 'cards recorded in the dataset')
```

```python
# How many females do we have information for vs how many males? 
males = df[df['Gender'] == 'Male']['Gender'].count()
females = df[df['Gender'] == 'Female']['Gender'].count()

print(f'We have information for {males} males in the dataset')
print(f'We have information for {females} females in the dataset')
```
